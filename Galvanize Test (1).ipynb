```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Reading the file
df = pd.read_csv('KNYC.csv')
```

```python
#Looking at the data frame, a few initial rows and what are its dimensions
print("The Shape of the data is : ", df.shape)
df.head(10)
```

```python
'''Looking at the details of different columns and some descriptive statistics'''
df['date']= pd.to_datetime(df['date'])
df.info()
df.describe()
```

It can be observed that standard deviation of "actual precipitaion" and "average_precipitation" is very small. This means that these variables do not change much. 

```python
#looking at the null, NAN values and duplicate rows in the dataset.
print("number of null values in the columns are:\n",df.isnull().sum())
print("number of NaN values in the columns are:\n",df.isna().sum())
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

We can see that there are no null values, no NaN values and no duplicate rows. Our data set is pretty clean.

```python
'''Making Box plots to check the outliers and general distribution of the data. 
Dividing the variables as per their scales'''
sns.boxplot(data=df.iloc[:,1:8], orient="h", palette="Set2")
```

```python
sns.boxplot(data=df.iloc[:,8:10], orient="h", palette="Set2")
```

We can see that there are no outliers in the above depicted variables. We can see that at times there are long whiskers, but there is not much skewness.

```python
ax=sns.boxplot(data=df.iloc[:,-3:-2], orient="h", palette="Set2")
```

```python
ax=sns.boxplot(data=df.iloc[:,-2:-1], orient="h", palette="Set2")
```

```python
ax=sns.boxplot(data=df.iloc[:,-1:], orient="h", palette="Set2")
```

```python
# Making the date as index.
df2 = df.set_index('date')
df2["actual_max_temp"].plot(legend=True)
df2["actual_min_temp"].plot(legend=True)
```

```python
df2["average_max_temp"].plot(legend=True)
df2["average_min_temp"].plot(legend=True)
```

We can see that Maximum and Minimum temperatures (actual and average) over the year are kind of locked. They move together, increase and decrease together, and can be seen as highly correlated variables.

```python
'''Looking at the frequency of years with record min and record max temperatures. 
It may give us information about general temperature trend'''

df2[['record_min_temp_year','record_max_temp_year']].hist(figsize=(7,7), bins=20, alpha=0.5)
```

We can see that as the years are increasing from 1880 frequency of minimum temperature is reducing while frequnecy of max temperature year has increased over time. This can be seen as a sign of global warming phenomena.  

```python
#plotting correlation
plt.figure(figsize=(15,15))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm')
```

We can see that all the temperature variables are highly correlated with each other. They move together, meaninig they increase and decrease together. Precipitation has no impact on actual and record precipitations as they have correlations really close to 0.

```python
#Looking at the rainfall trend throughout the year
plt.figure(figsize=(20,5))
df2["actual_precipitation"].plot(legend=True)
df2["average_precipitation"].plot(legend=True)
df2["record_precipitation"].plot(legend=True)
```

It can be noted that there is no particular rainy season in NYC. average precipitation remains fairly constant throughout the year. rainfalls are a little more prominent from october to feburary and a lot of record rainfall values are really for that period.
